## Task 9: Create Dataset Summary Table\n

```python
def summarize_table(name, df):
    return {
        "table_name": name,
        "rows": df.shape[0],
        "columns": df.shape[1],
        "missing_values": df.isnull().sum().sum(),
        "duplicate_rows": df.duplicated().sum()
    }

summary = pd.DataFrame([
    summarize_table(name, df)
    for name, df in tables.items()
])

summary\n
```

```python
summary.to_csv("../reports/lab01_dataset_summary.csv", index=False)\n
```

## Task 10: Missing Value Analysis\n

```python
def missing_value_summary(df):
    missing_count = df.isnull().sum()
    missing_percent = (missing_count / len(df)) * 100
    
    result = pd.DataFrame({
        "missing_count": missing_count,
        "missing_percent": missing_percent
    })
    
    return result[result["missing_count"] > 0].sort_values(
        by="missing_percent",
        ascending=False
    )

missing_value_summary(orders)\n
```

```python
for name, df in tables.items():
    print("\n" + "=" * 80)
    print(f"Missing Values in {name}")
    print("=" * 80)
    display(missing_value_summary(df))\n
```

## Reflection Questions

1. **Why is it important to understand the dataset before building a Machine Learning model?**
   Understanding the dataset is crucial to avoid "garbage in, garbage out". It helps identify missing values, outliers, the distribution of features, and the relationships between them. It ensures that the model is built on solid, reliable data and that assumptions about the business context hold true.

2. **Which table appears to be the central table in the Olist dataset? Why?**
   The `orders` table is the central table because it connects customers, payments, reviews, and order items. Nearly all other tables link back to `orders` via `order_id`.

3. **What is the difference between `customer_id` and `customer_unique_id`?**
   `customer_id` is assigned per order (a single customer making 3 orders will have 3 different `customer_id`s), whereas `customer_unique_id` uniquely identifies the actual person across all their orders.

4. **Which columns appear to represent dates?**
   Columns like `order_purchase_timestamp`, `order_approved_at`, `order_delivered_carrier_date`, `order_delivered_customer_date`, and `order_estimated_delivery_date`.

5. **Which columns appear to represent categorical variables?**
   `order_status`, `payment_type`, `customer_state`, `customer_city`, `product_category_name`.

6. **Which columns contain missing values?**
   Based on the missing value analysis, `orders` has missing values in dates (`order_approved_at`, `order_delivered_carrier_date`, `order_delivered_customer_date`). `products` has missing values in categories and physical dimensions. `reviews` has missing values in `review_comment_title` and `review_comment_message`.

7. **Which tables can be joined using `order_id`?**
   `orders`, `order_items`, `payments`, and `reviews`.

8. **Which tables can be joined using `product_id`?**
   `order_items` and `products`.

9. **What are three potential Machine Learning problems that can be created from this dataset?**
   - Predicting whether a package will be delivered late based on location and date.
   - Predicting the review score a customer will give based on delivery performance.
   - Customer segmentation based on their purchase behavior.

10. **What ethical concerns may arise when using customer data for ML?**
    Privacy issues (identifying users based on location data), algorithmic bias (e.g., offering worse service or higher prices to certain regions), and lack of transparency on how their data is used.\n
